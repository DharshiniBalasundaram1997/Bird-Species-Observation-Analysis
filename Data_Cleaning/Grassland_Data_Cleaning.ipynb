Specify the file path:

```python
import pandas as pd
import os

grassland_dataset_file_path = "C:/Users/dhars/Downloads/Dhass/codeing/GUVI/2. MainBoot/4.Project_Code/Project2/DataSet/Bird_Monitoring_Data_GRASSLAND.XLSX"
```

Reads the first sheet by default:

```python
excel_data = pd.read_excel(grassland_dataset_file_path)
excel_data
```

Read the Excel file with multiple sheets:

```python
excel_file = pd.ExcelFile(grassland_dataset_file_path)
# excel_file

# Get all sheet names
sheet_names = excel_file.sheet_names
sheet_names
```

Read data from all sheets into a dictionary:

```python
sheets_dict = {}

for i in sheet_names:
    df = excel_file.parse(i)
    sheets_dict[i] = df
sheets_dict
```

Convert sheets_dict to a single DataFrame:

```python
# Example: Convert sheets_dict to a single DataFrame; this will add a new sheet column at the last.
combined_df = pd.concat(
    [df.assign(Sheet=sheet_name) for sheet_name, df in sheets_dict.items()],
    ignore_index=True
)
combined_df.head()
```

Saving the Grassland DataSet file:

```python
# Create the directory if it doesn't exist
output_dir = 'C:/Users/dhars/Downloads/Dhass/codeing/GUVI/2. MainBoot/4.Project_Code/Project2/Bird_Species_Observation/DataSet/UnCleanDataSet'
os.makedirs(output_dir, exist_ok=True)

# Now save the file
combined_df.to_csv(os.path.join(output_dir, 'UnCleaned_Grassland_DataSet.csv'), index=False)
```
